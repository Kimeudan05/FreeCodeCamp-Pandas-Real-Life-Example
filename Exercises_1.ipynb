### Add and calculate a new `Calculated_Date` column

Use `Day`, `Month`, `Year` to create a `Date` column (`YYYY-MM-DD`).


```python
# your code goes here

# convert tthe month name to a crrect month numbers
sales['Month_N'] = pd.to_datetime(sales['Month'],format='%B').dt.month
# Construct full datetime from Year, Month_N, and Day
sales['Calculated_Date'] = pd.to_datetime(sales[['Year', 'Month_N', 'Day']].rename(columns={
    'Year':'year',
    'Month_N':'month',
    'Day':'day'
}))

sales.head()
```

```python
sales['Calculaated_Date'] = sales[['Year', 'Month', 'Day']].apply(lambda x: '{}-{}-{}'.format(x[0], x[1], x[2]), axis=1)

sales['Calculaated_Date'].head()
```
